# 1.5.9 Feature Matching

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('byc1.jpg',0) # queryImage
img2 = cv2.imread('byc2.jpg',0) # trainImage
```

```python
sift = cv2.SIFT_create()
kp1, des1 = sift.detectAndCompute(img1,None)
kp2, des2 = sift.detectAndCompute(img2,None)
```

```python
img1v=cv2.drawKeypoints(img1,kp1,img1,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
img2v=cv2.drawKeypoints(img2,kp2,img2,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.figure(figsize=(30,26))
plt.subplot(1,2,1), plt.imshow(img1v)
plt.subplot(1,2,2), plt.imshow(img2v)
plt.show()
```

```python
cv2.imwrite('byc1.jpg',img1v)
```

```python
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2, k=2)
# Apply ratio test
good = []
```

```python
for m,n in matches:
    if m.distance < 0.75*n.distance:
        good.append([m])
```

```python
print(good)
```

```python
img3 = cv2.drawMatchesKnn(img1,kp1,img2,kp2,good,None,flags=2)
```

```python
plt.figure(figsize=(30,15))
plt.imshow(img3),plt.show()
```

```python
cv2.imwrite('byc1.jpg',img3)
```
